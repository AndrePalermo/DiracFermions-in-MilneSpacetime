```python
import matplotlib.pyplot as plt
from matplotlib import rc
from plot_params import *
import numpy as np


linestyles = ["solid","dashed","dotted","dashdot","solid","dashed","dotted","dashdot"]



plt.rcParams['text.usetex'] = True


plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Computer Modern Roman']  


plt.rcParams['lines.linewidth'] = 3
plt.rcParams['font.size'] = 16
```

# Polarization plots

```python
## LONGITUDINAL POLARIZATION
folder = "./polarization_data"
filenames = ["Polarization_T=0.15_pT=0.txt","Polarization_T=0.3_pT=0.txt","Polarization_T=1.0_pT=0.txt"]
temps = ["0.15 GeV","0.30 GeV","1.0 GeV"]


plt.figure()
for i,f in enumerate(filenames):
    splist, pollist = np.loadtxt(f"{folder}/{f}")
    plt.plot(splist,pollist, label=f"T={temps[i]}",linestyle = linestyles[i])

plt.xlabel("$\Omega$[GeV]",fontsize=18)
plt.ylabel("$P_z$",fontsize=18)    
plt.legend(frameon=False)
plt.savefig("./longitudinal_pol.pdf",bbox_inches='tight')
```

```python
## POLARIZATION CHANGING PT
folder = "./polarization_data"
filenames = ["Polarization_T=0.15_pT=0.txt","Polarization_T=0.15_pT=0.1.txt","Polarization_T=0.15_pT=0.5.txt","Polarization_T=0.15_pT=1.0.txt"]
pTs = ["0 GeV (Analytic)","0.1 GeV","0.5 GeV","1.0 GeV"]

plt.figure()
for i,f in enumerate(filenames):
    splist, pollist = np.loadtxt(f"{folder}/{f}")
    plt.plot(splist,pollist, label=f"$p_T$={pTs[i]}",linestyle = linestyles[i])

plt.xlabel("$\Omega$[GeV]",fontsize=18)
plt.ylabel("$P_z$",fontsize=18)    
plt.legend(frameon=False)
plt.savefig("./longitudinal_polvspT.pdf",bbox_inches='tight')
```

```python
folder = "./txt_integrated"

linestyles = ["-","--"]

# =========================
# ENERGY
# =========================
plt.figure()

data = np.loadtxt(f"{folder}/energy_ratio.txt")
x = data[:,0]
y = data[:,1]

plt.plot(x, y, linestyle="-")
plt.axhline(1.0, color="grey", linewidth=3.0)

plt.xlabel(r"$\Omega$[GeV]", fontsize=18)
plt.ylabel(r"$\mathcal{E}_{\rm Can}/\mathcal{E}_{\rm Bel}$", fontsize=18)
plt.legend(frameon=False)
plt.savefig("./energy_ratio.pdf", bbox_inches='tight')


# =========================
# PRESSURES (T + L)
# =========================
plt.figure()

data_T = np.loadtxt(f"{folder}/transverse_pressure_ratio.txt")
data_L = np.loadtxt(f"{folder}/longitudinal_pressure_ratio.txt")

x = data_T[:,0]

plt.plot(x, data_T[:,1], label=r"$\mathcal{P}_{\mathrm{T}}$", linestyle=linestyles[0])
plt.plot(x, data_L[:,1], label=r"$\mathcal{P}_{\mathrm{L}}$", linestyle=linestyles[1])
plt.axhline(1.0, color="grey", linewidth=3.0)

plt.xlabel(r"$\Omega$[GeV]", fontsize=18)
plt.ylabel(r"$\mathcal{P}_{\rm Can}/\mathcal{P}_{\rm Bel}$", fontsize=18)
plt.legend(frameon=False)
plt.savefig("./pressures.pdf", bbox_inches='tight')


# =========================
# SPIN DENSITY
# =========================
plt.figure()

data = np.loadtxt(f"{folder}/spin_density.txt")
x = data[:,0]
y = data[:,1]

plt.plot(x, y, linestyle="-")

plt.xlabel(r"$\Omega$[GeV]", fontsize=18)
plt.ylabel(r"$\mathcal{S}\,[\mathrm{GeV}^3]$", fontsize=18)
plt.legend(frameon=False)
plt.savefig("./spin_density.pdf", bbox_inches='tight')


# =========================
# SPIN TORQUE
# =========================
plt.figure()

data = np.loadtxt(f"{folder}/spin_torque.txt")
x = data[:,0]
y = data[:,1]

plt.plot(x, y, linestyle="-")

plt.xlabel(r"$\Omega$[GeV]", fontsize=18)
plt.ylabel(r"$\mathcal{T}\,[\mathrm{GeV}^4]$", fontsize=18)
plt.legend(frameon=False)
plt.savefig("./spin_torque.pdf", bbox_inches='tight')

# =========================
# ANISOTROPY FACTOR
# =========================
plt.figure()

data = np.loadtxt(f"{folder}/anisotropy_factor.txt")
x = data[:,0]
y = data[:,1]

plt.plot(x, y, linestyle="-")
plt.axhline(1.0, color="grey", linewidth=3.0)

plt.xlabel(r"$\Omega$[GeV]", fontsize=18)
plt.ylabel(r"$\mathcal{P}_{\mathrm{L}}/\mathcal{P}_{\mathrm{T}}$", fontsize=18)
plt.legend(frameon=False)
plt.savefig("./anisotropy_factor.pdf", bbox_inches='tight')
```
