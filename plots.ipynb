```python
import matplotlib.pyplot as plt
from matplotlib import rc
from plot_params import *
import numpy as np


linestyles = ["solid","dashed","dotted","dashdot","solid","dashed","dotted","dashdot"]



plt.rcParams['text.usetex'] = True


plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Computer Modern Roman']  


plt.rcParams['lines.linewidth'] = 3
plt.rcParams['font.size'] = 16
```

# Polarization plots

```python
## LONGITUDINAL POLARIZATION
folder = "./polarization_data"
filenames = ["Polarization_T=0.15_pT=0.txt","Polarization_T=0.3_pT=0.txt","Polarization_T=1.0_pT=0.txt"]
temps = ["0.15 GeV","0.30 GeV","1.0 GeV"]


plt.figure()
for i,f in enumerate(filenames):
    splist, pollist = np.loadtxt(f"{folder}/{f}")
    plt.plot(splist,pollist, label=f"T={temps[i]}",linestyle = linestyles[i])

plt.xlabel("$\Omega$[GeV]",fontsize=18)
plt.ylabel("$P_z$",fontsize=18)    
plt.legend(frameon=False)
plt.savefig("./longitudinal_pol.pdf",bbox_inches='tight')
```

```python
## POLARIZATION CHANGING PT
folder = "./polarization_data"
filenames = ["Polarization_T=0.15_pT=0.txt","Polarization_T=0.15_pT=0.1.txt","Polarization_T=0.15_pT=0.5.txt","Polarization_T=0.15_pT=1.0.txt"]
pTs = ["0 GeV (Analytic)","0.1 GeV","0.5 GeV","1.0 GeV"]

plt.figure()
for i,f in enumerate(filenames):
    splist, pollist = np.loadtxt(f"{folder}/{f}")
    plt.plot(splist,pollist, label=f"$p_T$={pTs[i]}",linestyle = linestyles[i])

plt.xlabel("$\Omega$[GeV]",fontsize=18)
plt.ylabel("$P_z$",fontsize=18)    
plt.legend(frameon=False)
plt.savefig("./longitudinal_polvspT.pdf",bbox_inches='tight')
```
